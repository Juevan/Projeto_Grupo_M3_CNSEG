##Organizando os valores para o google sheets conseguir interpretar

##Arredondando valores, ajeitando o sinal de float, convertendo tipos

```python
df['High'] = df['High'].astype(str)
df['Low'] = df['Low'].astype(str)
df['Open'] = df['Open'].astype(str)
df['Close'] = df['Close'].astype(str)
df['Volume'] = df['Volume'].astype(str)


df['High'].str.replace(';', '')
df['Low'].str.replace(',', '.')
df['Open'].str.replace(',', '.')
df['Close'].str.replace(',', '.')
df['Volume'].str.replace(',', '.')

df['High'] = df['High'].astype(float).round(4)
df['Low'] = df['Low'].astype(float).round(4)
df['Open'] = df['Open'].astype(float).round(4)
df['Close'] = df['Close'].astype(float).round(4)
df['Volume'] = df['Volume'].astype(float).round(4)

df
```

##Tirar campos nulos


```python
null_rows = df[df.isnull().any(axis=1)]

print(null_rows)
```
